## Imports

```python
import numpy as np
import tensorflow as tf
from tf_agents.bandits.environments import stationary_stochastic_py_environment as sspe
from tf_agents.policies import random_py_policy
```

## Baseline Context Sampling Function

```python
def patient_context_sampling_fn():
    return np.random.rand(1, 2)
```

## Reward Function

```python
class MedicationReward(object):
    def __init__(self, theta):
        self.theta = theta

    def __call__(self, x):
        mu = np.dot(x, self.theta)
        return np.random.normal(mu, 1)
```

```python
reward_functions = [
    MedicationReward(theta) for theta in [(0.5, 1.5), (1.5, 0.5), (1, 1), (1, -1), (-0.5, -1.5), (-1.5, -0.5)]
]
```

## Environment

```python
env = sspe.StationaryStochasticPyEnvironment(patient_context_sampling_fn, reward_functions)
```

## Policy

```python
random_policy = random_py_policy.RandomPyPolicy(time_step_spec=env.time_step_spec(), action_spec=env.action_spec())
```

```python
for _ in range(100):
    time_step = env.reset()
    action = random_policy.action(time_step).action
    next_time_step = env.step(action)
    print("Context:", time_step.observation)
    print("Chosen Medication-Dosage:", action)
    print("Reward (Speech Fluency Improvement):", next_time_step.reward)
    print("------")
```
